### Kết hợp dữ liệu sử dụng hàm concat

```python
import wget
import os

urls = ["https://raw.githubusercontent.com/Sakaggi/dataengineeringtutorial/main/Transform/electricity_access_percent.csv",
        "https://raw.githubusercontent.com/Sakaggi/dataengineeringtutorial/main/Transform/rural_population_percent.csv"]

def download_url(urls):
    for url in urls:
        file_name = url.split("/")[-1]
        file_path = "{}".format("D:\@@\P\TransformData\\"+file_name)
        check_file = os.path.exists(path=file_path)
        if check_file:
            print(f"File existed :{file_path}")
        else:
            try:
                file_name = wget.download(url= url, out= "D:\@@\P\TransformData")
                print(f"File downloaded :{file_path}")
            except Exception as e:
                print(f"An error occured: {e}")
download_url(urls)
```

# Combine two file

```python
import pandas as pd

df_electrict = pd.read_csv("electricity_access_percent.csv", on_bad_lines='skip', skiprows=2, header= 1)
df_electrict = df_electrict.drop(['Unnamed: 62'], axis= 1)
df_electrict.head()
```

```python
df_rural = pd.read_csv("rural_population_percent.csv", on_bad_lines= "skip", skiprows=2, header=1)
df_rural = df_rural.drop(['Unnamed: 62'], axis= 1)
df_rural.head()
```

```python
# pd.set_option("display.max_rows", None)
df_concat = pd.concat([df_electrict, df_rural])
df_concat.tail()
```

# Combine the two data in the CSV file together so that the output looks like:

```python
# sample
l = [{"Country Name": "Aruba", "Country Code": "ABW", "Year": 1960, "Rural_value": 49.224, "Electricity_Value": 49.239}]

df_resutlt = pd.DataFrame(l)
df_resutlt
```

```python
# Using melt() to change formatting of each data frame
df_rural_melt = pd.melt(df_rural, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name="Year", value_name='Rural_value')
df_electrict_melt = pd.melt(df_electrict, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name="Year", value_name="Electrict_value")

# Using drop() to remove any columns from the data frames that aren't needed
df_rural_melt.drop(['Indicator Name', 'Indicator Code'], axis= 1, inplace= True)
df_electrict_melt.drop(['Indicator Name', 'Indicator Code'], axis= 1, inplace= True)

# using merge() to combine two dfs based on their common columns, and sort_values() to sort values by countruy and then by year.
df_merge = df_rural_melt.merge(df_electrict_melt, on= ['Country Name', 'Country Code', 'Year']).sort_values(['Country Name', 'Year'])
df_merge
```

```python
import wget
import os

urls = ["https://raw.githubusercontent.com/Sakaggi/dataengineeringtutorial/main/Extract/projects_data.csv", 
        "https://raw.githubusercontent.com/Sakaggi/dataengineeringtutorial/main/Extract/population_data.csv",
        "https://raw.githubusercontent.com/Sakaggi/dataengineeringtutorial/main/Extract/population_data.json",
        "https://raw.githubusercontent.com/Sakaggi/dataengineeringtutorial/main/Extract/population_data.xml"
        ]
def download_csv(urls):
    for url in urls:
        filename = url.split("/")[-1]
        output_path = "D:\@@\P"
        path = "{}".format(output_path+'\\'+filename)
        check_file = os.path.exists(path)
        if check_file:
            print(f"File downloaded: {output_path+filename}")
        else:
            try:
                filename = wget.download(url, out=output_path)
                print(f"File downloaded: {filename}")
            except Exception as e:
                print(f"An error occurred: {e}")
download_csv(urls)
```
